```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# CSV 파일 불러오기
df = pd.read_csv('/content/CarPrice.csv')  # 경로 상황에 따라 조정
df.columns = df.columns.str.strip()  # 공백 제거

# 이미지 불러오기
for idx, row in df.iterrows():
    car_name = row['CarName']
    img_filename = car_name + '.jpg'
    img_path = os.path.join('/content/images', img_filename)

    if os.path.exists(img_path):
        img = Image.open(img_path)
        plt.imshow(img)
        plt.title(car_name)
        plt.axis('off')
        plt.show()
    else:
        print(f"이미지 없음: {img_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# CSV 로드
df = pd.read_csv('/content/drive/MyDrive/2025-1/dl/cclass.csv')  # 또는 '/mnt/data/cclass.csv'
df.columns = df.columns.str.strip()  # 열 이름 공백 제거

# 이미지 경로 설정 및 시각화
for idx, row in df.iterrows():
    # 예: "C Class_2020.jpg" 형식
    file_name = f"{row['model']}_{row['year']}.jpg"
    file_path = os.path.join('/content/images', file_name)

    if os.path.exists(file_path):
        img = Image.open(file_path)
        plt.imshow(img)
        plt.title(file_name)
        plt.axis('off')
        plt.show()
    else:
        print(f"❌ 이미지 없음: {file_path}")
```
